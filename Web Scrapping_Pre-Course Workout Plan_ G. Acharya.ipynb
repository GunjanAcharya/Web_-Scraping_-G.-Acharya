## Importing required libraries:

```python
import numpy as np
import pandas as pd

from urllib.request import urlretrieve,urlopen,Request
import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings("ignore")
```

```python
url='https://hmidarjeeling.com/pre-course-workout-plan/'
```

```python
# Making a GET request:

r = requests.get('https://hmidarjeeling.com/pre-course-workout-plan/')

# check status code for response received: success code - 200
print(r)
```

```python
# Parsing the HTML:

soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify())
```

```python
# Getting the title tag 
print(soup.title) 
  
# Getting the name of the tag 
print(soup.title.name) 
  
# Getting the name of parent tag 
print(soup.title.parent.name) 
```

```python
# Finding Elements by class:

s=soup.find('span', {'style':"color: #ff6600;"})

  
for i in s:
    print(i.text)
```

```python
# Finding Elements by class:

s=soup.find('span', {'style':"color: #ff6600;"})
s=s.find_next('span', {'style':"color: #ff6600;"})
  
for i in s:
    print(i.text)
```

```python
# Finding Elements by class:

s=soup.find('span', {'style':"color: #ff6600;"})
s=s.find_next('span', {'style':"color: #ff6600;"})
s=s.find_previous('span', {'style':"color: #ff6600;"})

for i in s:
    print(i.text)
```

```python
# Finding Elements by class:

s=soup.find('span', {'style':"color: #ff6600;"})
print(s.text)
print('\n')
s=s.find_all_next('span', {'style':"color: #ff6600;"})
  
for i in s:
    print(i.text)
```

```python
# Finding Elements by class:

s=soup.find_all('p')
for i in s:
    print(i.text)
    print('\n')
```

```python
s=soup.find_all('td',{'width':'155'})

for i in s:
    print(i.text)
    print('\n')


```

```python
# Arranging the data sequentially:

t='1.      Warm Up. 2.      Free Squats (15 repetitions, 2 sets).3.      Single Leg Standing Calf Raise (12 repetitions, 1 set)4.      Inverted Rows (8 repetitions, 2 sets).5.      Push Ups / Regressed Push Ups (8 repetitions, 2 sets).6.      Prone Alternate Arm and Leg Raise. (10 repetitions on each side, 2 sets)7.      Partial Crunches (8 Repetitions, 2 sets).8.      Grip and Hang (till failure 1 set).9.      Cool Down.'
t2=''

for i in range(len(t)):
    if t[i].isdigit()==False:
        t2+=t[i]
        
    else:
        if t[i+1]=='.':
            
            t2+='\n'
            t2+=t[i]
        else:
            t2+=t[i]
print(t2)    
```

### Let's Scrape the WorkOut Plan for all the 12 Weeks:

```python
s=soup.find_all('td')

for i in s:
    print(i.text)
    print('\n')


```

```python
G={}

for i in ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']:
    G[i]=[]

for i in range(len(s)):
    if s[i].text not in ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']:
        continue
    else:

        G[s[i].text].append(s[i+7].text)

            
workout=pd.DataFrame(G,index = ['Week_'+str(i) for i in range(1,13)])
workout.head
```

```python
workout.Monday['Week_1']
```

```python
for i in workout:
    workout[i]=workout[i].str.replace('\xa0', ' ')
workout.Monday['Week_1']
```
