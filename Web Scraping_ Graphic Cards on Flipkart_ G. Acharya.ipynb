```python
description=[]
price=[]
highlights=[]
seller=[]
seller_rating=[]


for i in links:
    r = requests.get(i)
    soup = BeautifulSoup(r.content, 'html.parser')

    try:
        s2=soup.find('div',{'class':'_1mXcCf RmoJUa'}).text
        description.append(s2)
    except:
        description.append('N/A')

    try:
        s2=soup.find('div',{'class':'_30jeq3'}).text
        price.append(s2)
    except:
        price.append('N/A')

    try:
        s2=soup.find_all('li',{'class':'_21Ahn-'})
        highlights.append([])
        for j in s2:
            highlights[-1].append(j.text)
    except:
        highlights.append('N/A')

    try:
        s2=soup.find('div',{'class':'_1RLviY'}).text
        seller.append(s2)
    except:
        seller.append('N/A')

    try:
        s2=soup.find('div',{'class':'_3LWZlK _1D-8OL'}).text
        seller_rating.append(s2)
    except:
        seller_rating.append('N/A')
```

```python
print(len(description))
print(len(price))
print(len(highlights))
print(len(seller))
print(len(seller_rating))
```
